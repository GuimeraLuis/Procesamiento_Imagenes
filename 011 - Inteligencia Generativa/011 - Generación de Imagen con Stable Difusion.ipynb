```python
import cv2
import numpy as np
from diffusers import StableDiffusionXLPipeline
import torch
from PIL import Image
import gradio as gr
```

```python
#2.2 Cargar el modelo y preparar el dispositivo
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16
)
pipe.to("cuda")
```

```python
#2.3 Definir la función de generación
def generar_imagen(prompt):
    images = pipe(prompt, num_inference_steps=50, guidance_scale=7.5).images
    return images[0]
```

```python
#2.4 Crear la interfaz con Gradio
gr.Interface(
    fn=generar_imagen,
    inputs=gr.Textbox(label="Prompt para la imagen"),
    outputs=gr.Image(type="pil", label="Imagen generada"),
    title="Generador de Imágenes con Stable Diffusion XL",
    description="Escribí una descripción y generá una imagen con IA usando Stable Diffusion XL."
).launch()
```
